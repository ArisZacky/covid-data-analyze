# Kondisi umum

```python
sns.lineplot(x="Total Pasien", y="Sembuh", data=data).set(title="Pasien Sembuh")
```

```python
sns.lineplot(x="Total Pasien", y="Meninggal", data=data).set(title="Pasien Meninggal")
```

```python
sns.lineplot(x="Tanggal", y="Positif Harian", data=data).set(title="Positif Harian")
```

```python
date_sembuh = data.groupby(['Total Pasien'])[['Sembuh']].sum()
date_meninggal = data.groupby(['Total Pasien'])[['Meninggal']].sum()
perbandingan = date_sembuh.join(date_meninggal)

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
perbandingan.plot(ax=plt.gca(), title='Perbandingan')
```

```python
data_sembuh = data.groupby(['Total Pasien'])[['Sembuh']].sum()
data_meninggal = data.groupby(['Total Pasien'])[['Meninggal']].sum()

total_sembuh = data_sembuh['Sembuh'].values[-1:]
total_meninggal = data_meninggal['Meninggal'].values[-1:]

# Label pada sumbu x
labels = ['Data Pasien']

# Mengatur posisi pada sumbu x
x = np.arange(len(labels))

# Lebar bar
width = 0.35

# Membuat figure dan axes
fig, ax = plt.subplots()

# Plot bar untuk value1
rects1 = ax.bar(x - width/2, total_sembuh, width, label='Sembuh')

# Plot bar untuk value2
rects2 = ax.bar(x + width/2, total_meninggal, width, label='Meninggal')

# Menambahkan label pada sumbu x
ax.set_xticks(x)
ax.set_xticklabels(labels)

# Menambahkan label pada sumbu y
ax.set_ylabel('Banyak Pasien')

# Menambahkan judul
ax.set_title('Bar Chart: Sembuh vs Meninggal')

# Menambahkan legenda
ax.legend()

# Menampilkan bar chart
plt.show()
```

```python
# Membuat figure dan axes
fig, ax = plt.subplots()

# Plot bar untuk SumbuX1
ax.bar(data['Tanggal'], data['Meninggal'], width=0.4, align='center', label='Sembuh')

# Plot bar untuk SumbuX2
ax.bar(data['Tanggal'], data['Meninggal'], width=0.4, align='edge', label='Meninggal')

# Menambahkan label pada sumbu x
# ax.set_xticklabels(df.index.strftime('%Y-%m-%d'), rotation=45)

# Menambahkan label pada sumbu y
ax.set_ylabel('Banyak Pasien')

# Menambahkan judul
ax.set_title('Bar Chart: Sembuh vs Meninggal')

# Menambahkan legenda
ax.legend()

# Menampilkan bar chart
plt.tight_layout()
plt.show()
```
